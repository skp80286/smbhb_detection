```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/CRTS_Markarian_509.csv')

plt.figure().set_figwidth(14)
plt.gca().invert_yaxis()
plt.errorbar(df["MJD"], df["Mag"], yerr=df["Magerr"], fmt='o', ms=3, mfc='red', lw=0.5)
plt.savefig('output/CRTS_Markarian_509.png')
```
